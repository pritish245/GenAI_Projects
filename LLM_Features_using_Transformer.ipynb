# **Summarization**
  * **There are two primary kind of summarization**
    * **Extractive: This approach involves the identification & extraction of key phrases, sentences and segment from the original text and then these all would be combined to form a summary**
    * **Abstractive: This process is a bit different from abstractive, since it requires a more deeper understanding and generation capabilities**

* **Extractive Model**

```python
from transformers import pipeline
```

```python
summ_pipe = pipeline("summarization", model="facebook/bart-large-cnn")
```

```python
text = """
The tower is 324 metres (1,063 ft) tall, about the same height as an 81-storey building,
and the tallest structure in Paris. Its base is square, measuring 125 metres (410 ft) on each side.
During its construction, the Eiffel Tower surpassed the Washington Monument to become the
tallest man-made structure in the world, a title it held for 41 years until the Chrysler Building
in New York City was finished in 1930. It was the first structure to reach a height of 300 metres.
Due to the addition of a broadcasting aerial at the top of the tower in 1957,
 it is now taller than the Chrysler Building by 5.2 metres (17 ft). Excluding transmitters,
 the Eiffel Tower is the second tallest free-standing structure in France after the Millau Viaduct.
"""
```

```python
res = summ_pipe(text)
```

```python
res[0]["summary_text"]
```

* **Abstractive Model**

```python
abs_pipe = pipeline("text2text-generation", model="Atharvgarg/distilbart-xsum-6-6-finetuned-bbc-news-on-abstractive")
```

```python
res = abs_pipe(text, max_length = 130)
```

```python
res[0]["generated_text"]
```

# **Sentence Similarity**
  * **Why is there a need of sentence similarity?**
    * **Database: When we are working with an application, there is a need of memory. Data is in numerical format. VectorDB, is a kind of database that stores data in the form of vector**

```python
text = {
    "similar": [
        "The soft wind moved the flowers in the quiet garden",
        "A gentle breeze touched the flowers in the calm garden"
    ],

    "irrelevant" : [
        "The astronaut trained hard for the space trip",
        "The old castle stood on the hill above the small town"
    ],
}
```

```python
from sentence_transformers import SentenceTransformer, util
```

```python
model = SentenceTransformer("sentence-transformers/all-MiniLM-L6-v2")
```

```python
emb1 = model.encode(text["similar"][0], convert_to_tensor = True)
emb2 = model.encode(text["similar"][1], convert_to_tensor = True)
```

```python
emb1, emb2
```

```python
util.pytorch_cos_sim(emb1, emb2)
```

* **API Inference**

```python
from google.colab import userdata
```

```python
api_key = userdata.get("HFToken")
```

```python
API_URL = "https://api-inference.huggingface.co/models/sentence-transformers/msmarco-distilbert-base-tas-b"
headers = {"Authorization": f"Bearer {api_key}"}

def query(payload):
	response = requests.post(API_URL, headers=headers, json=payload)
	return response.json()
```

```python
import requests
```

```python
data = query(
    {
        "inputs" : {
            "source_sentence" :text["similar"][0], # Sentence, which is going to be compared
            "sentences" : [ # Comparsion
                text["similar"][1],
                text["irrelevant"][0],
                text["irrelevant"][1],
                ]
        }
    }
)
```

# **Text - to - Speech (TTS)**

```python
text = """Hi, My name is Dev, and I work as a Researcher at Intellipaat. We have been working with Generative AI for fine-tuning
operaitons. Moreover, maths is fun"""
```

```python
from IPython.display import Audio
```

```python
suno_pipe = pipeline(model = "suno/bark-small")
```

```python
output = suno_pipe(text) # Array
```

```python
audio_array = output["audio"]
```

```python
sample_rate = 22000
```

```python
Audio(audio_array, rate = sample_rate)
```

**Facebook**

```python
facebook_pipe = pipeline(model = "facebook/mms-tts-eng")
```

```python
output = facebook_pipe(text)
```

```python
Audio(output["audio"], rate = output["sampling_rate"])
```
